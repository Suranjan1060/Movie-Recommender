```python
# Importing necessary libraries
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings('ignore')

# Loading the dataset
data_path = '/content/drive/MyDrive/IMDBAnalysis-main/IMDBAnalysis-main/archive/IMDB-Movie-Data.csv'
data = pd.read_csv(data_path)

# Cleaning the column names
data.columns = data.columns.str.strip()

# Task 1: Find the Top 10 Rated Movies in the Comedy Genre
comedy_movies = data[data['Genre'].str.contains('Comedy', case=False)]
top_10_comedy = comedy_movies.nlargest(10, 'Rating')[['Title', 'Rating', 'Director']].set_index('Title')
print("\nTop 10 Rated Comedy Movies:")
print(top_10_comedy)

# Visualizing the top 10 rated comedy movies
plt.figure(figsize=(10, 5))
sns.barplot(x=top_10_comedy['Rating'], y=top_10_comedy.index)
plt.title("Top 10 Rated Comedy Movies")
plt.xlabel("Rating")
plt.ylabel("Movie Title")
plt.show()

# Task 2: Find the Top 5 Rated Movies Overall
top_5_movies = data.nlargest(5, 'Rating')[['Title', 'Rating', 'Director']].set_index('Title')
print("\nTop 5 Rated Movies Overall:")
print(top_5_movies)

# Visualizing the top 5 rated movies overall
plt.figure(figsize=(10, 5))
sns.barplot(x=top_5_movies['Rating'], y=top_5_movies.index)
plt.title("Top 5 Rated Movies Overall")
plt.xlabel("Rating")
plt.ylabel("Movie Title")
plt.show()
```
